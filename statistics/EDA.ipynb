## cleaning and filtering the data

</span>

```python
# finding the missing values
boat.isnull()
```

```python
# finding the number of missing values in column
boat.isnull().sum()
```

```python
# to drop those column with large missing values
boat_clean=boat.drop(["deck"],axis=1)
boat_clean
```

```python
boat_clean.isnull().sum()
```

```python
cleaned_dataset=boat_clean.dropna()
```

```python
cleaned_dataset.isnull().sum()
```
